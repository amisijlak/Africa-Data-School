### Convert numerical variable to  categorical variable

##### Bin
- Binning is a way to group a number of more or less continuous values into a smaller number of "bins".
- Binning method is used to smoothing data or to handle noisy data. 
- In this method, the data is first sorted and then the sorted values are distributed into a number of buckets or bins. 
- The original data values which fall into a given small interval, a bin, are replaced by a value representative of that interval, often the central value.

```python
print(df.head())
print(df['SepalLengthCm'].min())
print(df['SepalLengthCm'].max())
```

```python
sepal_length_ranges = [f"{SepalLengthCm + 0.1} - {SepalLengthCm + 1}" for SepalLengthCm in range(4, 8, 1)]
sepal_length_ranges
```

```python
count_sepal_length_ranges = len(sepal_length_ranges)
count_sepal_length_ranges
```

```python
#df['SepalLengthCm_Range'] = pd.cut(x=df['SepalLengthCm'], bins=count_sepal_length_ranges, labels=sepal_length_ranges)
df.head()
```

### Binning Function

```python
# Define the function which take two argument, the range of bins as rng and the 'SepalLengthCm' as x column.
def binSepals(rng,x):
    # loop through the bins and split each of them to obtain the upper and lower boundaries for each bin.
    for i in rng:
        lower = i.split(' - ')[0]
        upper = i.split(' - ')[1]
        # Check the range in which each value of the of SepalLengthCm lies.
        if float(x) <= float(upper) and float(x)  >= float(lower):
            # return the bin.
            return f'({i})'
                
                
df['SepalLengthCm_Range'] = df['SepalLengthCm'].apply(lambda x: binSepals(sepal_length_ranges,x))
```

```python
df
```

```python
sepal_len_hist_df = df['SepalLengthCm_Range'].value_counts()
sepal_len_hist_df
```

```python
df_range_hist = pd.DataFrame(sepal_len_hist_df).reset_index()
df_range_hist.columns = ['SepalLengthCm_Range', 'Count']
df_range_hist
```

```python
plt.bar(df_range_hist["SepalLengthCm_Range"], df_range_hist["Count"])
plt.show()
```

It's a common tool for describing simple relationships without making a statement about the effect.

A negative correlation, or inverse correlation, is a key concept in the creation of diversified portfolios that can better withstand portfolio volatility.

#### Chi Square Test of Independence.
Test Relationships between categorical variables (contingency tables).
The chi-squared statistic is a single number that tells you how much difference exists between your observed counts and the counts you would expect if there were no relationship at all in the population. ... A low value for chi-square means there is a high correlation between your two sets of data.

A chi-square test for independence compares two variables in a contingency table to see if they are related. In a more general sense, it tests to see whether distributions of categorical variables differ from each another.

A chi square test will give you a p-value. The p-value will tell you if your test results are significant or not

```python
Image(filename="chi.png")
```

```python
voter_race = np.random.choice(a = ['asian','black', 'hispanic', 'other', 'white'],
                              p = [0.05, 0.15, 0.25, 0.05, 0.5],
                              size=100
                             )
voter_party = np.random.choice(a = ['democrat', 'independent', 'republican'],
                               p = [0.4, 0.2, 0.4],
                               size=100
                              )
voters = pd.DataFrame({'race': voter_race, 'party': voter_party})
voters.head()
```

```python
voter_tab = pd.crosstab(voters.race, voters.party, margins=True)
voter_tab.columns = ['democrat', 'independent', 'republican', 'row_total']
voter_tab.index = ['asian','black', 'hispanic', 'other', 'white', 'column_total']
voter_tab
```

```python
chi2, p, dof, ex = scipy.stats.chi2_contingency(voter_tab)
print(chi2)
print(p)
print(dof)
print(ex)
```

## Numerical-Categorical Variable Analysis

#### one-way ANOVA
- The one-way ANOVA tests whether the mean of some numeric variable differs across the levels of one categorical variable.
- It essentially answers the question: do any of the group means differ from one another?
- If there are two groups you can also  used two-sample t-test instead of ANOVA.

- H<sub>0</sub>: All means are the same.
- H<sub>a</sub>: Atleast one differs.
- Significance level: 5%;

```python
races =   ["asian","black","hispanic","other","white"]

# Generate random data
voter_race = np.random.choice(a= races,
                              p = [0.05, 0.15 ,0.25, 0.05, 0.5],
                              size=100)

voter_age = scipy.stats.poisson.rvs(loc=18,
                              mu=30,
                              size=100)

# Group age data by race
voter_frame = pd.DataFrame({"race":voter_race,"age":voter_age})
groups = voter_frame.groupby("race").groups
# Extract individual groups
asian = voter_age[groups["asian"]]
black = voter_age[groups["black"]]
hispanic = voter_age[groups["hispanic"]]
other = voter_age[groups["other"]]
white = voter_age[groups["white"]]
# Perform the ANOVA
scipy.stats.f_oneway(asian, black, hispanic, other, white)
```

- Anova test explains whether there is significant difference between the variables, but doesn't show which variables are different.
- To get which variables are different, we perform a Tukey's Range Test to determine the difference.
- Tukey's range Test is performed using pairwise_turkeyhsd.
- It will return a table of information telling you whether or not there is difference for each pair of variables.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
tukey = pairwise_tukeyhsd(endog=voter_age,     # Data
                          groups=voter_race,   # Groups
                          alpha=0.05)          # Significance level

tukey.summary()              # See test summary
```
